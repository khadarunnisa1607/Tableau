Q7. Explain the business importance of geographical analysis.

Then describe how you would create a Map showing Sales by State using Sample – Superstore.

Hint:  Use State and Sales. Tableau automatically detects geographic fields.

ANS:-

Geographical analysis helps businesses understand sales performance across different locations such as states or regions. It helps identify high-performing and low-performing areas, improve marketing strategies, optimize resource allocation, and support better business decisions.

How to create a Map showing Sales by State in Sample – Superstore
Open a new worksheet.

Drag State to the worksheet. Tableau automatically recognizes it as a geographic field and creates a map.

Drag Sales to the Color shelf on the Marks card (or to Size to vary the size of the marks).

The map will display the sales for each state, with color intensity (or size) representing the total sales.

This allows you to compare sales performance across different states and quickly identify regions with higher or lower sales.

Q8. What is the difference between a Worksheet and a Dashboard in Tableau?

Why is a dashboard more useful for decision-makers?


ANS:-| **Worksheet**                                         | **Dashboard**                                                              |
| ----------------------------------------------------- | -------------------------------------------------------------------------- |
| A worksheet contains a single chart or visualization. | A dashboard combines multiple worksheets and visualizations on one screen. |
| It is used to analyze one aspect of the data.         | It is used to present a complete overview of the data.                     |
| Each worksheet displays one view.                     | A dashboard displays multiple views together.                              |



Why is a Dashboard more useful for decision-makers?

A dashboard provides a comprehensive view of important business information in one place. It allows decision-makers to monitor key metrics, compare different visualizations, identify trends, and make faster, data-driven decisions without switching between multiple worksheets.


Q9.  What is Tableau Prep, and why is it important in real-world data projects?

What is Tableau Prep, and why is it important in real-world data projects?

Then explain how you would:

Remove null values

Remove duplicate records

Change incorrect data types (Answer in steps, no tool access required.)




ANS:-

**Tableau Prep** is a data preparation tool used to clean, combine, and organize data before analyzing it in Tableau. It helps improve data quality by fixing errors and preparing data for accurate analysis.

**Importance:**

* Cleans messy data.
* Improves data accuracy.
* Saves time during analysis.
* Ensures reliable and meaningful insights.

### a) Remove Null Values

**Steps:**

1. Open the dataset in Tableau Prep.
2. Identify fields that contain null values.
3. Select the field and filter out **Null** values or replace them with appropriate values.
4. Verify that the null values have been removed.

### b) Remove Duplicate Records

**Steps:**

1. Load the dataset into Tableau Prep.
2. Identify duplicate rows using the **Clean** step.
3. Remove the duplicate records.
4. Verify that only unique records remain.

### c) Change Incorrect Data Types

**Steps:**

1. Open the **Clean** step.
2. Select the field with the incorrect data type.
3. Click the data type icon next to the field name.
4. Change it to the correct type (such as **String, Number, Date,** or **Boolean**).
5. Verify that the field is displayed correctly.
