## Q9. Chart Selection Justification

Different charts were used because each chart is best suited for answering a specific business question.

* **Bar Chart (Sales by Category):** Used to compare sales across product categories and identify the highest-selling category.
* **Line Chart (Monthly Sales Trend):** Used to analyze sales changes over time and identify growth trends, peak months, and low-performing periods.
* **Pie Chart (Region-wise Sales Contribution):** Used to show the proportion of total sales contributed by each region.
* **Scatter Plot (Sales vs Profit Analysis):** Used to analyze the relationship between sales and profit and identify profitable and loss-making orders.
